# Download data

```python
# Define the start and end dates for the data
start = datetime.datetime(2000, 2, 2)
end = datetime.datetime(2023, 8, 31)
```

```python
# Fetch data for various indicators
oil_prices = yf.download('CL=F', start=start, end=end)  # Crude Oil prices
gold_prices = yf.download('GC=F', start=start, end=end)  # Gold prices
nasdaq = yf.download('^IXIC', start=start, end=end)  # NASDAQ Composite Index
djia = yf.download('^DJI', start=start, end=end)     # Dow Jones Industrial Average
vix = yf.download('^VIX', start=start, end=end)      # CBOE Volatility Index
```

```python
oil_prices
```

```python
extra_daily_data = pd.DataFrame(index=daily_range)
```

```python
extra_daily_data
```

```python
def fill_dataframe(data, dl_df, col_name1, col_name2):
  first_index = dl_df.index[0]
  last_index = dl_df.index[-1]

  # Ensure the indices match for the assignment
  common_index = data.loc[first_index:last_index].index.intersection(dl_df.index)

  # Assign values using .loc for DateTime indexing
  data.loc[common_index, col_name1] = dl_df.loc[common_index, col_name2]

  return data
```

```python
extra_daily_data = fill_dataframe(extra_daily_data, oil_prices, 'oil', 'Close')
extra_daily_data = fill_dataframe(extra_daily_data, gold_prices, 'gold', 'Close')
extra_daily_data = fill_dataframe(extra_daily_data, nasdaq, 'nasdaq', 'Close')
extra_daily_data = fill_dataframe(extra_daily_data, djia, 'djia', 'Close')
extra_daily_data = fill_dataframe(extra_daily_data, vix, 'vix', 'Close')
```

```python
# Economic data from FRED
interest_rate = web.DataReader('FEDFUNDS', 'fred', start, end)  # Federal Funds Rate
inflation_rate = web.DataReader('CPIAUCSL', 'fred', start, end)  # Consumer Price Index
gdp_growth = web.DataReader('GDP', 'fred', start, end)           # Gross Domestic Product
unemployment_rate = web.DataReader('UNRATE', 'fred', start, end) # Unemployment Rate
```

```python
# Convert data to daily frequency
interest_rate_daily = interest_rate.resample('D').asfreq().fillna(np.nan)
inflation_rate_daily = inflation_rate.resample('D').asfreq().fillna(np.nan)
gdp_growth_daily = gdp_growth.resample('D').asfreq().fillna(np.nan)
unemployment_rate_daily = unemployment_rate.resample('D').asfreq().fillna(np.nan)
```

```python
interest_rate_daily.interpolate(method='time', inplace=True)
inflation_rate_daily.interpolate(method='time', inplace=True)
gdp_growth_daily.interpolate(method='time', inplace=True)
unemployment_rate_daily.interpolate(method='time', inplace=True)
```

```python
# Exchange Rates
usd_eur = web.DataReader('DEXUSEU', 'fred', start, end)  # USD to EUR
usd_jpy = web.DataReader('DEXJPUS', 'fred', start, end)  # USD to JPY
usd_gbp = web.DataReader('DEXUSUK', 'fred', start, end)  # USD to GBP
usd_cny = web.DataReader('DEXCHUS', 'fred', start, end)  # USD to CNY
usd_cad = web.DataReader('DEXCAUS', 'fred', start, end)  # USD to CAD
usd_mxn = web.DataReader('DEXMXUS', 'fred', start, end)  # USD to MXN
```

```python
extra_daily_data.loc[subset.index, 'usd_eur'] = usd_eur['DEXUSEU']
extra_daily_data.loc[subset.index, 'usd_jpy'] = usd_jpy['DEXJPUS']
extra_daily_data.loc[subset.index, 'usd_gbp'] = usd_gbp['DEXUSUK']
extra_daily_data.loc[subset.index, 'usd_cny'] = usd_cny['DEXCHUS']
extra_daily_data.loc[subset.index, 'usd_cad'] = usd_cad['DEXCAUS']
extra_daily_data.loc[subset.index, 'usd_mxn'] = usd_mxn['DEXMXUS']
```

```python
extra_daily_data
```

```python
extra_daily_data['oil'].fillna(method='bfill', inplace=True)
extra_daily_data['gold'].fillna(method='bfill', inplace=True)

extra_daily_data['nasdaq'].interpolate(method='time', inplace=True)
extra_daily_data['djia'].interpolate(method='time', inplace=True)
extra_daily_data['vix'].interpolate(method='time', inplace=True)

extra_daily_data['usd_eur'].interpolate(method='time', inplace=True)
extra_daily_data['usd_jpy'].interpolate(method='time', inplace=True)
extra_daily_data['usd_gbp'].interpolate(method='time', inplace=True)
extra_daily_data['usd_cny'].interpolate(method='time', inplace=True)
extra_daily_data['usd_cad'].interpolate(method='time', inplace=True)
extra_daily_data['usd_mxn'].interpolate(method='time', inplace=True)
```
