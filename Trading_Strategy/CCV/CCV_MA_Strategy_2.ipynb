# Strategy

## Without volatility

```python
def strategy(data, short_window, long_window, risk_free_rate=0.04, initial_capital=1000):

  data['Short_MA'] = data['Close'].rolling(window=short_window).mean()
  data['Long_MA'] = data['Close'].rolling(window=long_window).mean()

  data['equity'].iloc[0] = initial_capital
  account_balance = initial_capital
  # profit_loss = []
  total_shares = 0
  trades = []
  position = 'none'
  buy = 0
  sell = 0
  last_close = 0

  for i in range(1, len(data)):
    if (data['Short_MA'].iloc[i] >= data['Long_MA'].iloc[i] and position == 'none'):
      # buy
      buy_shares = (account_balance) / data['Close'].iloc[i]
      account_balance -= buy_shares * data['Close'].iloc[i]
      data['position'].iloc[i] = 'buy'
      total_shares += buy_shares
      position = 'buy'
      buy = 1
      data['equity'].iloc[i] = account_balance + (total_shares * data['Close'].iloc[i])
      trades.append(('buy', data.index[i], data['Close'].iloc[i], data['equity'].iloc[i]))

    elif (data['Short_MA'].iloc[i] <= data['Long_MA'].iloc[i] and buy == 1):
      # close buy
      account_balance += buy_shares * data['Close'].iloc[i]
      data['position'].iloc[i] = 'close buy'
      total_shares -= buy_shares
      buy_shares = 0
      position = 'none'
      buy = 0
      last_close = i
      data['equity'].iloc[i] = account_balance + (total_shares * data['Close'].iloc[i])
      trades.append(('close buy', data.index[i], data['Close'].iloc[i], data['equity'].iloc[i]))

    else:
      data['equity'].iloc[i] = account_balance + (total_shares * data['Close'].iloc[i])

  # Calculate returns
  data['Daily Return'] = data['equity'].pct_change()
  data['Strategy Return'] = data['Daily Return'].shift(1)


  # Calculate Sharpe Ratio
  excess_returns = data['Strategy Return']
  sharpe_ratio = ((excess_returns.mean() - (risk_free_rate / 252)) / excess_returns.std()) * np.sqrt(252)

  # Calculate total profit
  total_profit = data['equity'].iloc[last_close] - initial_capital

  # profit_loss = []
  # for i in range(1, len(data)):
  #   pl = data['Daily Return'].iloc[i] * data['equity'].iloc[i-1]
  #   profit_loss.append(pl)
  # total_profit = builtins.sum(profit_loss)

  # Calculate drawdown
  data['Cumulative Max'] = data['equity'].cummax()
  data['Drawdown'] = data['equity'] - data['Cumulative Max']
  max_drawdown = data['Drawdown'].min()
  max_drawdown_percentage = (max_drawdown / data['Cumulative Max'].max()) * 100

  return total_profit, sharpe_ratio, max_drawdown_percentage, trades
```

```python
def backtest_function(strategy, data, volatility, short_window, long_window):
  results = pd.DataFrame(columns=['Volatility', 'Total profit', 'Sharpe ratio', 'Max drawdown (%)'])

  for vol in volatility:
    data_f = data
    total_profit, sharpe_ratio, max_drawdown_percentage, trades = strategy(data_f, data_f[vol], short_window, long_window)
    new_row = {'Volatility': vol, 'Total profit': total_profit, 'Sharpe ratio': sharpe_ratio, 'Max drawdown (%)': max_drawdown_percentage}
    results.loc[len(results)] = new_row

  return results
```

```python
def check_strategy(data, CCV):

    sharpe_ratio_CCV = data.loc[data['Volatility'] == CCV, 'Sharpe ratio'].values[0]
    sharpe_ratio_pure = data.loc[data['Volatility'] == '__', 'Sharpe ratio'].values[0]

    max_drawdown_CCV = data.loc[data['Volatility'] == CCV, 'Max drawdown (%)'].values[0]
    max_drawdown_pure = data.loc[data['Volatility'] == '__', 'Max drawdown (%)'].values[0]

    total_profit_CCV = data.loc[data['Volatility'] == CCV, 'Total profit'].values[0]
    total_profit_pure = data.loc[data['Volatility'] == '__', 'Total profit'].values[0]

    is_CCV_sharpe_positive = sharpe_ratio_CCV > 0
    is_CCV_sharpe_greater_than_pure_sharpe = sharpe_ratio_CCV > sharpe_ratio_pure
    is_CCV_dd_greater_than_pure_dd = max_drawdown_CCV > max_drawdown_pure
    is_pure_tp_greater_than_CCV_tp = total_profit_pure > total_profit_CCV

    all_sharpes_greater_than_Pure_sharpe = (data.loc[(data['Volatility'] != '__') & (data['Volatility'] != 'CCV') & (data['Volatility'] != 'CCV_dw') & (data['Volatility'] != 'CCV_cw'), 'Sharpe ratio'] > sharpe_ratio_pure)
    all_sharpes_greater_than_Pure_sharpe_percentage = (all_sharpes_greater_than_Pure_sharpe.sum()/8) * 100

    positive_sharpes = (data.loc[(data['Volatility'] != '__') & (data['Volatility'] != 'CCV') & (data['Volatility'] != 'CCV_dw') & (data['Volatility'] != 'CCV_cw'), 'Sharpe ratio'] > 0)
    positive_sharpes_percentage = (positive_sharpes.sum()/8) * 100

    CCV_sharpe_greater_than_all = (data.loc[(data['Volatility'] != '__') & (data['Volatility'] != 'CCV') & (data['Volatility'] != 'CCV_dw') & (data['Volatility'] != 'CCV_cw'), 'Sharpe ratio'] < sharpe_ratio_CCV)
    CCV_sharpe_greater_than_all_percentage = (CCV_sharpe_greater_than_all.sum()/8) * 100

    CCV_dd_greater_than_all = (data.loc[(data['Volatility'] != '__') & (data['Volatility'] != 'CCV') & (data['Volatility'] != 'CCV_dw') & (data['Volatility'] != 'CCV_cw'), 'Max drawdown (%)'] < max_drawdown_CCV)
    CCV_dd_greater_than_all_percentage = (CCV_dd_greater_than_all.sum()/8) * 100

    CCV_tp_greater_than_all = (data.loc[(data['Volatility'] != '__') & (data['Volatility'] != 'CCV') & (data['Volatility'] != 'CCV_dw') & (data['Volatility'] != 'CCV_cw'), 'Total profit'] < total_profit_CCV)
    CCV_tp_greater_than_all_percentage = (CCV_tp_greater_than_all.sum()/8) * 100

    return is_CCV_sharpe_positive, is_CCV_sharpe_greater_than_pure_sharpe, is_CCV_dd_greater_than_pure_dd, all_sharpes_greater_than_Pure_sharpe_percentage, positive_sharpes_percentage, CCV_sharpe_greater_than_all_percentage, CCV_dd_greater_than_all_percentage, CCV_tp_greater_than_all_percentage
```
